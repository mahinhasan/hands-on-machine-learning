```python
index = count()

x_val = []
y_val = []

def animate(i):
    x_val.append(next(index))
    y_val.append(random.randint(0,5))
    plt.cla()
    plt.plot(x_val,y_val)
    plt.show()

ani = FuncAnimation(plt.gcf(),animate,interval=10000)
plt.tight_layout()
plt.show()
```
